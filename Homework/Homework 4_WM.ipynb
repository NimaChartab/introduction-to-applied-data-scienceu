```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt 
```

```python
Data = pd.read_csv("LendingClub.csv")
Data
```

```python
Data.dtypes
```

```python
Data.isnull().sum()
```

```python
Data["term"] = Data["term"].astype('category')
Data["term"] = Data["term"].cat.codes

Data["grade"] = Data["grade"].astype('category')
Data["grade"] = Data["grade"].cat.codes

Data["sub_grade"] = Data["sub_grade"].astype('category')
Data["sub_grade"] = Data["sub_grade"].cat.codes

Data["home_ownership"] = Data["home_ownership"].astype('category')
Data["home_ownership"] = Data["home_ownership"].cat.codes

Data["verification_status"] = Data["verification_status"].astype('category')
Data["verification_status"] = Data["verification_status"].cat.codes


Data["loan_status"] = Data["loan_status"].astype('category')
Data["loan_status"] = Data["loan_status"].cat.codes


Data["purpose"] = Data["purpose"].astype('category')
Data["purpose"] = Data["purpose"].cat.codes

Data["earliest_cr_line"] = Data["earliest_cr_line"].astype('category')
Data["earliest_cr_line"] = Data["earliest_cr_line"].cat.codes


Data["initial_list_status"] = Data["initial_list_status"].astype('category')
Data["initial_list_status"] = Data["initial_list_status"].cat.codes


Data["application_type"] = Data["application_type"].astype('category')
Data["application_type"] = Data["application_type"].cat.codes


Data["address"] = Data['address'].str[-5:]


Data
```

```python
Data.dtypes
```

```python
X= Data.drop(['loan_status'],axis=1)
y=Data[['loan_status']]
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(X.corr(),annot=True)
plt.show()
```

```python
#look for visual correlations. Can only use this if I explicitly set X.

#sns.pairplot(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values,test_size =0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
#dense layers. intial layer is input layer. there are 22 inputs, so 22 neurons.

#hidden layers utilize all labels, so 23 neurons

#output layer, there is only 1 output so 1 neuron.

#problem is classification, so use sigmoid (binary). #use binary_crossentropy for classification

#MSE and MAE for regression usually.

model=Sequential()

model.add(Dense(22,activation='sigmoid'))

model.add(Dense(23,activation='sigmoid'))

model.add(Dense(23,activation='sigmoid'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) 
```

```python
n_features = X.shape[1]
n_classes = y.shape[1]
```

```python
X_train.shape
```

```python
model.fit(x=X_train, y=y_train,validation_data=(X_test,y_test),batch_size=32,epochs=20)

#smaller batch_size number, lower chance of overfitting
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
loss=pd.DataFrame(model.history.history)
loss.plot()
```

```python
predictions=model.predict_classes(X_test) #use model.predict for regression #model.predict_classes for classification
```

```python
predictions
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test,predictions)
cf

#[[TN,FP],[FN,TP]]
```

```python
#used for reference

TP,TN,FP,FN=0,0,0,0
for i in range(len(y_test)):
    if y_test[i]==1:
        if predictions[i]==1:
            TP+=1
        if predictions[i]==0:
            FN+=1
    if y_test[i]==0:
        if predictions[i]==1:
            FP+=1
        if predictions[i]==0:
            TN+=1
```

```python
print(np.array([[TP,FP],[FN,TN]]))
```

```python
precision=cf[1][1]/(cf[1][1]+cf[0][1]) #precision=TP/(TP+FP)
print("precision=",np.round(precision*100,2),"%")
```

```python

recall=cf[1][1]/(cf[1][1]+cf[1][0]) #recall=TP/(TP+FN)
print("recall=",np.round(recall*100,2),"%")
```

```python
F1_score=2*precision*recall/(precision+recall)
print("F1_score=",np.round(F1_score*100,2),"%")
```

```python


accuracy=(cf[1][1]+cf[0][0])/(cf[0][0]+cf[0][1]+cf[1][0]+cf[1][1]) 
#accuracy=(TP+TN)/(TP+FP+FN+TN)
print("accuracy=",accuracy)
```
